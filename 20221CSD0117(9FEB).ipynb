```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("Step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,5,100)
print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,5,100)
print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,5,100)

ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,5,100)

ax.plot(x,np.sin(x),label='sin',color='r')
ax.plot(x,np.cos(x),label='cos',color='g')
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,5,100)

ax.plot(x,np.sin(x),label='sin',color='r',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',color='g',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='sin x+1',color='g',linestyle='dashdot')
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```
